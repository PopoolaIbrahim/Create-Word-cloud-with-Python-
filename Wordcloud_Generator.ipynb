```python
 # Import required libraries
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a string of words
text = 'Python JavaScript Numpy Pandas Matplotlib Seaborn Keras Tensorflow Fastai Python Python Pytorch'

# Generate word cloud
wordcloud = WordCloud(width=500, height=500, background_color='white').generate(text)

# Show the generated image
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
# Save the image to a file
wordcloud.to_file("wordcloud.png")
```
